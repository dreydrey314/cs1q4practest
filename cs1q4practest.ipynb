```python
#G7 CS Practical Test For the 4th Quarter
name = input("Enter your Name:")
Qscore = int(input("Enter your Score:"))
Qscore2 = int(input("Enter your score for quiz 2:"))
Qscore3 = int(input("Enter your score for quiz 3:"))
#Calculate the total score
sum = Qscore + Qscore2 + Qscore3
print("Your total score is:",sum)

#Calculate the average
avg = sum / 3
print("Your average score is:",avg)

#Determine whether the student passed or failed
if avg >= 50:
    print("You passed the exam")
else:
    print("You failed the exam")


```
